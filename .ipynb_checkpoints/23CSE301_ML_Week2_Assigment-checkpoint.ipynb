<span style = "font-size : 30px; color:Brown "> <b>Exercise-2</b> </span>

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

<span style = "font-size : 20px; color:Brown "> <b>1. Distribution Plot</b> </span>

```python
sns.histplot(data = iris , x = 'sepal_width', bins = 20)
plt.title("Histogram for Sepal Width")
plt.show()
```

<span style = "font-size : 20px; color:Brown "> <b>2. Categorical Plot</b> </span>

```python
sns.boxplot(data = iris , x = "species" , y= "sepal_width", color = "green")
plt.show()
```

<span style = "font-size : 20px; color:Brown "> <b>3. Relationship Plot</b> </span>

```python
sns.scatterplot(x = "sepal_length",y = "sepal_width",data = iris,hue = "species")
plt.title("Scatterplot")
plt.show()
```

<span style = "font-size : 20px; color:Brown "> <b>4. Correlation Heatmap</b> </span>

```python
corr = iris.drop('species' ,axis = 1).corr()
sns.heatmap(corr,annot = True , cmap="twilight")
plt.title("Correlation Heatmap")
plt.show()
print(plt.colormaps())
```

<span style = "font-size : 20px; color:Brown "> <b>5. Pairplot</b> </span>

```python
sns.pairplot(iris , hue = "species" )
plt.show()
```
